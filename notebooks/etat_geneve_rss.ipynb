```python
import feedparser


def afficher_donnees(data, niveau=0):
    indentation = "  " * niveau

    if isinstance(data, dict):
        for cle, valeur in data.items():
            print(f"{indentation}{cle}:")
            afficher_donnees(valeur, niveau + 1)

    elif isinstance(data, list):
        for index, element in enumerate(data):
            print(f"{indentation}[{index}]")
            afficher_donnees(element, niveau + 1)

    else:
        print(f"{indentation}{data}")

flux = feedparser.parse(url)

for index, article in enumerate(flux.entries, start=1):
    print(f"\n=== ARTICLE {index} ===")
    afficher_donnees(dict(article))
```

```python
import csv
import json
import re
from pathlib import Path
from typing import Optional

import requests
from bs4 import BeautifulSoup, NavigableString, Tag
from requests.adapters import HTTPAdapter
from urllib3.util.retry import Retry


def nettoyer_texte(value: str) -> str:
    """Normalise les espaces, tabulations et espaces insécables."""
    return re.sub(r"\s+", " ", value.replace("\xa0", " ")).strip()


def creer_session() -> requests.Session:
    """Crée une session HTTP avec nouvelles tentatives automatiques."""
    session = requests.Session()

    retry_strategy = Retry(
        total=3,
        connect=3,
        read=3,
        backoff_factor=1,
        status_forcelist=(429, 500, 502, 503, 504),
        allowed_methods=frozenset({"GET"}),
    )

    adapter = HTTPAdapter(max_retries=retry_strategy)
    session.mount("https://", adapter)
    session.mount("http://", adapter)

    session.headers.update(
        {
            "User-Agent": (
                "Mozilla/5.0 (Windows NT 10.0; Win64; x64) "
                "AppleWebKit/537.36 (KHTML, like Gecko) "
                "Chrome/126.0 Safari/537.36"
            ),
            "Accept-Language": "fr-CH,fr;q=0.9,en;q=0.8",
        }
    )

    return session


def trouver_index(
    lignes: list[str],
    libelle: str,
    debut: int = 0,
) -> Optional[int]:
    """Trouve la position d'un libellé dans une liste de textes."""
    libelle_normalise = nettoyer_texte(libelle).casefold()

    for index in range(debut, len(lignes)):
        ligne = lignes[index].casefold()

        if ligne == libelle_normalise:
            return index

        if ligne.startswith(f"{libelle_normalise} :"):
            return index

        if ligne.startswith(f"{libelle_normalise}:"):
            return index

    return None


def extraire_valeur_simple(
    lignes: list[str],
    libelle: str,
) -> Optional[str]:
    """
    Extrait une valeur située sur la même ligne que le libellé
    ou sur la ligne suivante.
    """
    index = trouver_index(lignes, libelle)

    if index is None:
        return None

    ligne = lignes[index]

    resultat = re.match(
        rf"^{re.escape(libelle)}\s*:\s*(.+)$",
        ligne,
        flags=re.IGNORECASE,
    )

    if resultat:
        return nettoyer_texte(resultat.group(1))

    if index + 1 < len(lignes):
        return lignes[index + 1]

    return None


def extraire_intervalle(
    lignes: list[str],
    libelle_debut: str,
    libelle_fin: str,
) -> Optional[str]:
    """Extrait les textes situés entre deux libellés."""
    index_debut = trouver_index(lignes, libelle_debut)

    if index_debut is None:
        return None

    ligne_debut = lignes[index_debut]

    resultat = re.match(
        rf"^{re.escape(libelle_debut)}\s*:\s*(.+)$",
        ligne_debut,
        flags=re.IGNORECASE,
    )

    valeurs: list[str] = []

    if resultat:
        valeurs.append(nettoyer_texte(resultat.group(1)))

    index_fin = trouver_index(
        lignes,
        libelle_fin,
        debut=index_debut + 1,
    )

    if index_fin is None:
        index_fin = len(lignes)

    valeurs.extend(lignes[index_debut + 1:index_fin])

    valeurs = [
        valeur
        for valeur in valeurs
        if valeur and valeur.casefold() != libelle_debut.casefold()
    ]

    return nettoyer_texte(" ".join(valeurs)) if valeurs else None


def extraire_article(contenu_principal: Tag) -> Optional[str]:
    """
    Extrait le corps de l'offre à partir de la section « Votre mission »
    jusqu'aux métadonnées « Type de publication ».
    """
    titre_depart = contenu_principal.find(
        lambda tag: (
            isinstance(tag, Tag)
            and tag.name in {"h2", "h3"}
            and nettoyer_texte(tag.get_text(" ", strip=True)).casefold()
            == "votre mission"
        )
    )

    if titre_depart is None:
        return None

    article: list[str] = []

    for element in titre_depart.next_elements:
        if isinstance(element, NavigableString):
            texte_brut = nettoyer_texte(str(element))

            if texte_brut.casefold() == "type de publication":
                break

            continue

        if not isinstance(element, Tag):
            continue

        if element.name not in {"h2", "h3", "h4", "p", "li"}:
            continue

        if element.name == "p" and element.find_parent("li") is not None:
            continue

        texte = nettoyer_texte(element.get_text(" ", strip=True))

        if not texte:
            continue

        if texte.casefold() == "type de publication":
            break

        if element.name in {"h2", "h3", "h4"}:
            ligne = f"## {texte}"
        elif element.name == "li":
            ligne = f"- {texte}"
        else:
            ligne = texte

        if not article or nettoyer_texte(article[-1]) != nettoyer_texte(ligne):
            article.append(ligne)

    resultat = "\n\n".join(article).strip()
    return resultat or None


def scraper_offre(url: str) -> dict[str, Optional[str]]:
    """Télécharge et extrait les informations de l'offre."""
    session = creer_session()

    response = session.get(url, timeout=(10, 30))
    response.raise_for_status()

    soup = BeautifulSoup(response.content, "html.parser")

    contenu_principal = (
        soup.find("main")
        or soup.find(attrs={"role": "main"})
        or soup.body
        or soup
    )

    lignes = [
        nettoyer_texte(texte)
        for texte in contenu_principal.stripped_strings
        if nettoyer_texte(texte)
    ]

    titre_element = contenu_principal.find("h1")
    titre = (
        nettoyer_texte(titre_element.get_text(" ", strip=True))
        if titre_element
        else None
    )

    return {
        "Titre": titre,
        "Date de publication": extraire_valeur_simple(
            lignes, "Date de publication"
        ),
        "Rémunération": extraire_valeur_simple(
            lignes, "Rémunération"
        ),
        "Taux d'activité": extraire_valeur_simple(
            lignes, "Taux d'activité"
        ),
        "Type de contrat": extraire_valeur_simple(
            lignes, "Type de contrat"
        ),
        "Délai d'inscription": extraire_valeur_simple(
            lignes, "Délai d'inscription"
        ),
        "Entrée en fonction": extraire_valeur_simple(
            lignes, "Entrée en fonction"
        ),
        "Article": extraire_article(contenu_principal),
        "Type de publication": extraire_intervalle(
            lignes,
            "Type de publication",
            "Auteur",
        ),
        "Auteur": extraire_intervalle(
            lignes,
            "Auteur",
            "Mise à jour",
        ),
        "Mise à jour": extraire_valeur_simple(
            lignes, "Mise à jour"
        ),
    }


def enregistrer_csv(
    donnees: dict[str, Optional[str]],
    fichier: str = "offre_emploi_geneve.csv",
) -> None:
    """Enregistre les données dans un fichier CSV."""
    chemin = Path(fichier)

    with chemin.open(
        "w",
        newline="",
        encoding="utf-8-sig",
    ) as fichier_csv:
        writer = csv.DictWriter(
            fichier_csv,
            fieldnames=list(donnees.keys()),
            delimiter=";",
        )
        writer.writeheader()
        writer.writerow(donnees)


def main() -> None:
    try:
        offre = scraper_offre(URL)

        print(json.dumps(offre, ensure_ascii=False, indent=2))

        enregistrer_csv(offre)
        print("\nFichier créé : offre_emploi_geneve.csv")

    except requests.RequestException as erreur:
        print(f"Erreur lors du téléchargement de la page : {erreur}")

    except Exception as erreur:
        print(f"Erreur pendant l'extraction : {erreur}")


if __name__ == "__main__":
    main()
```
